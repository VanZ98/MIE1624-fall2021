### c) Bootstrap mean distribution and mean difference

```python
#Separate out man and woman rows
df2_man = df2[df2['Q2']=='Man']
df2_woman = df2[df2['Q2']=='Woman']
#print(df2_man.Q2.unique())
```

```python
# sample male and female separately according to the size proportion
mean_boot_diffs = []
boot_man_mean = []
boot_woman_mean = []

for i in range(1000):
    bootstrap_sample_man = df2_man.sample(n = len(df2_man), replace = True)
    boot_man_mean.append(bootstrap_sample_man.Q24.mean())
    bootstrap_sample_woman = df2_woman.sample(n = len(df2_woman), replace = True)
    boot_woman_mean.append(bootstrap_sample_woman.Q24.mean())
    mean_boot_diff = bootstrap_sample_man.Q24.mean() - bootstrap_sample_woman.Q24.mean()
    mean_boot_diffs.append(mean_boot_diff)

#Create df for plotting
mean_boot_df = pd.DataFrame({'Mean Diff': mean_boot_diffs})
man_means = pd.DataFrame({'Man_Means': boot_man_mean})
woman_means = pd.DataFrame({'Woman_Means': boot_woman_mean})
```

```python
#plot bootstrap gender means distribution
plt.subplots(figsize=(13,8), dpi=100)
sns.distplot(man_means, color="blue", label="Male Group Distribution")
sns.distplot(woman_means, color="red", label="Female Group Distribution")
plt.title('Men vs. Women Bootstrap Means')
plt.tight_layout()
plt.xlabel('Salary Means', fontsize=12)
plt.ylabel('Frequency',fontsize=12)
plt.legend()
```

```python
#plot bootstrap mean difference distribution
plt.subplots(figsize=(13,8), dpi=100)
mean_boot_df = pd.DataFrame({'Mean Diff': mean_boot_diffs})
sns.distplot(mean_boot_df)
fig.tight_layout()
plt.title('Men vs. Women Bootstrap Means Differences')
plt.xlabel('Salary Mean Difference', fontsize=12)
plt.ylabel('Frequency',fontsize=12)
```

### d) Perform a two_sample t-test with a 0.05 threshold on bootstrapped sample means

According to CLT, the sampling distribution of the means follows a normal distribution, and t-test is then applied to the bootstrapped sample means to test the null hypothesis, with a threshold significance level of 0.05. The resulting p-value equals 0 tells that the difference between the salary means of male and female is statistically significant (<0.05), and the null hypothesis is therefore rejected.

```python
#perfrom two sample t-test on bootstrapped gender means
from scipy import stats
tc_bs, pc_bs = stats.ttest_ind(man_means, woman_means)
print ("t-test: t = %g  p = %g" % (tc_bs, pc_bs))
```

### Perform ANOVA test if applicable:

```python
# Shapiro-Wilks test to check ANOVA assumptions:
from scipy.stats import shapiro
alpha = 0.05

#bachelor
stat_b, p_b = shapiro(df3_b.Q24)
print('Bachelor degree: Statistics=%.3f, p=%.3f' % (stat_b, p_b))
# interpret
if p_b > alpha:
	print('Bachelor Sample looks Gaussian (fail to reject H0)')
else:
	print('Bachelor Sample does not look Gaussian (reject H0)')
print('\n')

#Master
stat_m, p_m = shapiro(df3_m.Q24)
print('Master degree: Statistics=%.3f, p=%.3f' % (stat_m, p_m))
# interpret
if p_m > alpha:
	print('Bachelor Sample looks Gaussian (fail to reject H0)')
else:
	print('Bachelor Sample does not look Gaussian (reject H0)')

print('\n')
#Doctoral   
stat_d, p_d = shapiro(df3_d.Q24)
print('Doctoral degree: Statistics=%.3f, p=%.3f' % (stat_d, p_d))
# interpret
if p_d > alpha:
	print('Bachelor Sample looks Gaussian (fail to reject H0)')
else:
	print('Bachelor Sample does not look Gaussian (reject H0)')
```

ANOVA test was used to compare the means of more than 2 groups. One of the ANOVA assumptions is that the residuals are approximately normally distributed. If the test results in a p-value greater than the threshold (alpha = 0.05), we can deduce that the dataset is very likely drawn from a Gaussian distribution and then apply the ANOVA test to compare the means. Upon applying the Shapiro-Wilks tests, a p-value of 0 was obtained for all three datasets, hence the ANOVA test is not applicable here. 

### Bootstarp the three groups separately and compute the pair-wise mean difference

```python
# bootstrap sample for normalized distribution
mean_boot_diffs_mb = []
mean_boot_diffs_md = []
mean_boot_diffs_bd = []

bs_bachelor_mean = []
bs_master_mean = []
bs_doctor_mean = []

for i in range(1000):
    
    bootstrap_sample_b = df3_b.sample(n = len(df3_b), replace = True)
    bs_bachelor_mean.append(bootstrap_sample_b.Q24.mean())
    bootstrap_sample_m = df3_m.sample(n = len(df3_m), replace = True)
    bs_master_mean.append(bootstrap_sample_m.Q24.mean())
    bootstrap_sample_d = df3_d.sample(n = len(df3_d), replace = True)
    bs_doctor_mean.append(bootstrap_sample_d.Q24.mean())
    
    #calc mean diff for each pair and append to the list
    mean_boot_diff_mb = bootstrap_sample_m.Q24.mean() - bootstrap_sample_b.Q24.mean()
    mean_boot_diffs_mb.append(mean_boot_diff_mb)
    mean_boot_diff_md = bootstrap_sample_d.Q24.mean() - bootstrap_sample_m.Q24.mean()
    mean_boot_diffs_md.append(mean_boot_diff_md)
    mean_boot_diff_bd = bootstrap_sample_d.Q24.mean() - bootstrap_sample_b.Q24.mean()
    mean_boot_diffs_bd.append(mean_boot_diff_bd)
    
mean_boot_Q4 = pd.DataFrame({'bachelor':bs_bachelor_mean,
                                   'master':bs_master_mean,
                                   'doctoral':bs_doctor_mean})

mean_boot_diffs_Q4 = pd.DataFrame({'M/B':mean_boot_diffs_mb,
                                   'M/D':mean_boot_diffs_md,
                                   'B/D':mean_boot_diffs_bd})
```

### Visualize the boostrap sample distribution and mean difference distribution

```python
#plot the salary means of bootstrapped samples
fig, ax = plt.subplots(figsize=(13,8), dpi=100)
ax.set_title('Mean Distribution by Education Level with Bootstrap', fontsize=15)
sns.distplot(mean_boot_Q4.bachelor, color="green", label='Bachelor’s degree')
sns.distplot(mean_boot_Q4.master , color="red", label='Master’s degree')
sns.distplot(mean_boot_Q4.doctoral , color="blue", label='Doctoral degree')
plt.tight_layout()
plt.xlabel('Salary_Means', fontsize=12)
plt.ylabel('Frequency',fontsize=12)
plt.legend()
```

```python
#Estimate of avg yearly salary for three groups based on bootstrap sampling means
mu, sigma = np.mean(mean_boot_Q4.bachelor), np.std(mean_boot_Q4.bachelor)
print ("Bachelor's degree avg annual salary = %G, st. dev = %g" % (mu, sigma))

mu, sigma = np.mean(mean_boot_Q4.master), np.std(mean_boot_Q4.master)
print ("Master's degree avg annual salary = %G, st. dev = %g" % (mu, sigma))

mu, sigma = np.mean(mean_boot_Q4.doctoral), np.std(mean_boot_Q4.doctoral)
print ("Doctoral degree avg annual salary = %G, st. dev = %g" % (mu, sigma))
```

```python
#plot the pair-wise mean differences for bootstraped samples
fig, ax = plt.subplots(1,1,figsize=(13,8), dpi=100)
ax.set_title('Mean Difference Distribution by Education Level with Bootstrap', fontsize=15)
sns.distplot(mean_boot_diffs_Q4['M/B'],bins=40, label = 'Mean_diff b/t Bachelor degree & Master degree')
sns.distplot(mean_boot_diffs_Q4['M/D'],bins=40, label = 'Mean_diff b/t Doctoral degree & Master degree')
sns.distplot(mean_boot_diffs_Q4['B/D'],bins=40, label = 'Mean_diff b/t Bachelor degree & Doctoral degree')
plt.tight_layout()
plt.xlabel('Salary_Mean_Difference', fontsize=12)
plt.ylabel('Frequency',fontsize=12)
plt.legend()
```

```python
# confidence interval
alpha = 0.05
left = np.percentile(mean_boot_diffs_mb, alpha/2*100)
right = np.percentile(mean_boot_diffs_mb, 100-alpha/2*100)
print((1-alpha)*100,'%','confidence interval for the difference between means - Bachelor vs. Master :', (round(left,2), round(right,2)))
#point estimate on the difference in mean
print("point estimate on difference in avg salaries - Bachelor vs. Master:", round(mean_boot_diffs_Q4['M/B'].mean(),2), '\n')

left = np.percentile(mean_boot_diffs_md, alpha/2*100)
right = np.percentile(mean_boot_diffs_md, 100-alpha/2*100)
print((1-alpha)*100,'%','confidence interval for the difference between means - Doctoral vs. Master :', (round(left,2), round(right,2)))
#point estimate on the difference in mean
print("point estimate on difference in avg salaries - Doctoral vs. Master:", round(mean_boot_diffs_Q4['M/D'].mean(),2), '\n')

left = np.percentile(mean_boot_diffs_bd, alpha/2*100)
right = np.percentile(mean_boot_diffs_bd, 100-alpha/2*100)
print((1-alpha)*100,'%','confidence interval for the difference between means - Doctoral vs. Bachelor :', (round(left,2), round(right,2)))
#point estimate on the difference in mean
print("point estimate on difference in avg salaries - Doctoral vs. Bachelor:", round(mean_boot_diffs_Q4['B/D'].mean(),2))
```
