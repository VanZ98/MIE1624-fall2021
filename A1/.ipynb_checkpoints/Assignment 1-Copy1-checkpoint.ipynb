### b) Perform a two_sample t-test with a 0.05 threshold

```python
gender_data = df2.groupby('Q2')
print(gender_data.mean())
```

As seen from above descriptive statistics for female and male groups, the means for two groups are quiet different. In order to evaluate whether the difference is statistically significant, the two sample t-test is performed as follow:

```python
from scipy import stats
male_salary = df2[df2['Q2']=='Man']['Q24'].astype('int')
female_salary = df2[df2['Q2']=='Woman']['Q24'].astype('int')
tc, pc = stats.ttest_ind(male_salary, female_salary)
print ("t-test: t = %g  p = %g" % (tc, pc))
```

Based on the t-test, it can be concluded that the difference between the average salary of male and female is significant (p = 4.77315e-15 < 0.05)

### c) Bootstrap mean distribution and mean difference

```python
def bootstrap_ci(df, variable, classes, repetitions = 1000, alpha = 0.05, random_state=None): 
    
    df = df[[variable, classes]]
    bootstrap_sample_size = len(df) 
    
    mean_diffs = []

    for i in range(repetitions):
        bootstrap_sample = df.sample(n = bootstrap_sample_size, replace = True, random_state = random_state)
        mean_diff = bootstrap_sample.groupby(classes).mean().iloc[1,0] - bootstrap_sample.groupby(classes).mean().iloc[0,0]
        mean_diffs.append(mean_diff)
    # confidence interval
    left = np.percentile(mean_diffs, alpha/2*100)
    right = np.percentile(mean_diffs, 100-alpha/2*100)
    # point estimate
    point_est = df.groupby(classes).mean().iloc[1,0] - df.groupby(classes).mean().iloc[0,0]
    print('Point estimate of difference between means:', round(point_est,2))
    print((1-alpha)*100,'%','confidence interval for the difference between means:', (round(left,2), round(right,2)))
```

```python
df_c = df2[['Q24','Q2']]
mean_diffs = []
bootstrap_sample_size = len(df2)
for i in range(1000):
    bootstrap_sample = df_c.sample(n = bootstrap_sample_size, replace = True, random_state = None)
    mean_diff = bootstrap_sample.groupby('Q2').mean().iloc[0,0] - bootstrap_sample.groupby('Q2').mean().iloc[1,0]
    mean_diffs.append(mean_diff)
print(bootstrap_sample)
```

```python
# sample male and female separately according to the size proportion
mean_boot_diffs = []
for i in range(1000):
    bootstrap_sample_man = df2_man.sample(n = len(df2_man), replace = True)
    bootstrap_sample_woman = df2_woman.sample(n = len(df2_woman), replace = True)
    mean_boot_diff = bootstrap_sample_man.Salary_Man.mean() - bootstrap_sample_woman.Salary_Woman.mean()
    mean_boot_diffs.append(mean_boot_diff)
```

### Bootstarp the three groups separately and compute the pair-wise mean difference

```python
# bootstrap sample for normalized distribution
mean_boot_diffs_mb = []
mean_boot_diffs_md = []
mean_boot_diffs_bd = []

for i in range(1000):
    
    bootstrap_sample_b = df3_b.sample(n = len(df3_b), replace = True)
    bootstrap_sample_m = df3_m.sample(n = len(df3_m), replace = True)
    bootstrap_sample_d = df3_d.sample(n = len(df3_d), replace = True)
    
    #calc mean diff for each pair and append to the list
    mean_boot_diff_mb = bootstrap_sample_m.Q24.mean() - bootstrap_sample_b.Q24.mean()
    mean_boot_diffs_mb.append(mean_boot_diff_mb)
    mean_boot_diff_md = bootstrap_sample_d.Q24.mean() - bootstrap_sample_m.Q24.mean()
    mean_boot_diffs_md.append(mean_boot_diff_md)
    mean_boot_diff_bd = bootstrap_sample_d.Q24.mean() - bootstrap_sample_b.Q24.mean()
    mean_boot_diffs_bd.append(mean_boot_diff_bd)
    
mean_boot_diffs_Q4 = pd.DataFrame({'M/B':mean_boot_diffs_mb,
                                   'M/D':mean_boot_diffs_md,
                                   'B/D':mean_boot_diffs_bd})
```

### Visualize the boostrap sample distribution and mean difference distribution

```python
plt.subplots(figsize=(13,8), dpi=100)
sns.distplot(bootstrap_sample_b.Q24 , color="green", label='Bachelor’s degree')
sns.distplot(bootstrap_sample_m.Q24 , color="red", label='Master’s degree')
sns.distplot(bootstrap_sample_d.Q24 , color="blue", label='Doctoral degree')
plt.tight_layout()
plt.xlabel('Salary', fontsize=12)
plt.ylabel('Frequency',fontsize=12)
plt.legend()
```

```python
fig, ax = plt.subplots(1,1,figsize=(13,8), dpi=100)
ax.set_title('Mean Difference Distribution by Education Level with Bootstrap', fontsize=15)
sns.distplot(mean_boot_diffs_Q4['M/B'],bins=40, label = 'Mean_diff b/t Bachelor degree & Master degree')
sns.distplot(mean_boot_diffs_Q4['M/D'],bins=40, label = 'Mean_diff b/t Doctoral degree & Master degree')
sns.distplot(mean_boot_diffs_Q4['B/D'],bins=40, label = 'Mean_diff b/t Bachelor degree & Doctoral degree')
plt.tight_layout()
plt.xlabel('Salary', fontsize=12)
plt.ylabel('Frequency',fontsize=12)
plt.legend()
```
